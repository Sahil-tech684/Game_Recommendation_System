# **Data Preprocessing**

```python
games_df.drop(columns=['website'],inplace=True)
games_df.dropna(subset=['about','genre'], inplace=True)
```

```python
games_df.duplicated().sum()
```

```python
def clean_string(x):
    return x.strip().replace(' ', '') if isinstance(x, str) else x
```

```python
columns_to_clean = ['publishers', 'genre', 'developers', 'tags', 'other_games_series']
for col in columns_to_clean:
    games_df[col] = games_df[col].apply(clean_string)

games_df
```

```python
games_df.about = games_df.about.apply(lambda x: [x])
games_df.genre = games_df.genre.apply(lambda x: [x])
games_df.publishers = games_df.publishers.apply(lambda x: [x])
games_df.developers = games_df.developers.apply(lambda x: [x])
games_df.tags = games_df.tags.apply(lambda x: [x])
games_df.other_games_series = games_df.other_games_series.apply(lambda x: [x])
```

```python
games_df['info'] = games_df.about + games_df.genre + games_df.publishers + games_df.developers + games_df.tags + games_df.other_games_series
```

```python
games_df['info'] = games_df['info'].apply(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)
games_df['info'] = games_df['info'].replace(np.nan, '', regex=True)
games_df['info'] = games_df['info'].str.replace("'", "").str.replace(',',' ').str.split(', ').apply(lambda x: ' '.join(x))
games_df['info'] = games_df['info'].str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
games_df.iloc[0]['info']
```

```python
games_df = games_df[['id','name','info','poster','about']]
```

```python
games_df['info'] = games_df['info'].apply(lambda x: x.lower())
games_df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []

    for i in text.split():
        x = ps.stem(i)
        y.append(x)

    return " ".join(y)
```

```python
games_df['info'] = games_df['info'].apply(stem)
```

```python
games_df.iloc[0]['info']
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
games_df['info'] = games_df['info'].apply(remove_stopwords)
```

```python
games_df.iloc[0]['info']
```

# **Vectorization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

tfidf_vectors = tfidf_vectorizer.fit_transform(games_df['info'])
tfidf_vectors_dense = tfidf_vectors.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_vectors_dense)
similarity
```

```python
def recommend(movie):
    index = games_df[games_df['name'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:11]:
        print(games_df.iloc[i[0]]['name'])
```

```python
recommend('Grand Theft Auto V')
```
